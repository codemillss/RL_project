#### 시각화를 위한 전처리 진행

```python
# Transforming the data using melt
jeju_2017_long = jeju_2017.melt(id_vars=['거래일자'], var_name='시간', value_name='수요량')

# Converting '시간' from string to integer after removing '시'
jeju_2017_long['시간'] = jeju_2017_long['시간'].str.replace('시', '').astype(int)

# Sorting the data by '거래일자' and '시간' and resetting the index
jeju_2017_long = jeju_2017_long.sort_values(by=['거래일자', '시간']).reset_index(drop=True)

# Display the transformed data
jeju_2017_long
```

```python
# Transforming the data using melt
jeju_2023_long = jeju_2023.melt(id_vars=['거래일자'], var_name='시간', value_name='수요량')

# Converting '시간' from string to integer after removing '시'
jeju_2023_long['시간'] = jeju_2023_long['시간'].str.replace('시', '').astype(int)

# Sorting the data by '거래일자' and '시간' and resetting the index
jeju_2023_long = jeju_2023_long.sort_values(by=['거래일자', '시간']).reset_index(drop=True)

# Display the transformed data
jeju_2023_long
```

```python
# Transforming the data using melt
jeju_2024_long = jeju_2024.melt(id_vars=['거래일자'], var_name='시간', value_name='수요량')

# Converting '시간' from string to integer after removing '시'
jeju_2024_long['시간'] = jeju_2024_long['시간'].str.replace('시', '').astype(int)

# Sorting the data by '거래일자' and '시간' and resetting the index
jeju_2024_long = jeju_2024_long.sort_values(by=['거래일자', '시간']).reset_index(drop=True)

# Display the transformed data
jeju_2024_long
```

# 제대로 형변환 된 것을 확인

```python
import matplotlib.pyplot as plt

# Create the datetime for plotting
jeju_2017_long['Datetime'] = pd.to_datetime(jeju_2017_long['거래일자']) + pd.to_timedelta(jeju_2017_long['시간'] - 1, unit='h')

# Plotting the demand over time
plt.figure(figsize=(14, 7))
plt.plot(jeju_2017_long['Datetime'], jeju_2017_long['수요량'], label='수요량', color='blue')
plt.xlabel('시간')
plt.ylabel('수요량 (Demand)')
plt.title('시간에 따른 전력 수요량 (Electricity Demand Over Time)')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()  # Adjust layout to make room for the rotated x-axis labels
plt.show()
```

```python
"""
데이터 구조에 문제가 없어보이니 
2024 데이터를 기준으로 잡고 나머지 데이터에 1000을 나눠주어  
w-> mw 단위로 통일 시켜줍니다.
"""
```

```python
jeju_2017_long['수요량'] = jeju_2017_long['수요량'] / 1000
jeju_2023_long['수요량'] = jeju_2023_long['수요량'] / 1000

```

```python
# jeju_2017_long, jeju_2023_long, jeju_2024_long 데이터 프레임이 이미 정의되어 있고 같은 구조를 가지고 있다고 가정
all_data = pd.concat([jeju_2017_long, jeju_2023_long, jeju_2024_long], axis=0)

# 필요하다면 인덱스를 리셋할 수 있습니다
all_data.reset_index(drop=True, inplace=True)

# 합쳐진 데이터 프레임의 첫 몇 줄을 출력하여 확인
print(all_data.head())
```
